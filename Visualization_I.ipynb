```python
temp = df2['origin'].value_counts().reset_index() #Untuk mendapatkan jumlah total dari masing2 kategori di kolom origin.
```

```python
temp.head()
```

```python
plt.bar(temp['index'], temp['origin'], color=['red', 'green', 'blue'])
####### X -Axis  ######   Y-Axis ##### Warna Untuk tiap kategori Index
```

```python
df1['sepal_width_cat'] = np.where(df1['sepal_width'] < 2.5, '< 2.5', #Kategori 1
                         np.where(df1['sepal_width'] < 3, '2.5 - 3', #Kategori 2
                         np.where(df1['sepal_width'] < 3.5, '3 - 3.5', #Kategori 3
                         np.where(df1['sepal_width'] < 4, '3.5 - 4', '> 4')))) #Kategori 4

temp2 = df1['sepal_width_cat'].value_counts().reset_index() #Hitung jumlah per kategori

temp2['index'] = pd.Categorical(temp2['index'], ordered=True,
                   categories=['< 2.5', '2.5 - 3', '3 - 3.5', '3.5 - 4', '> 4'])
```

```python
temp2
```

```python
plt.bar(temp2['index'], temp2['sepal_width_cat'])
```

```python
from pandas.core.internals.blocks import CategoricalDtype
df1['sepal_length_cat']=np.where(df1.sepal_length < 5,'< 5',
                        np.where(df1.sepal_length < 6,'5-5.9',
                        np.where(df1.sepal_length < 7,'6-6.9',
                        np.where(df1.sepal_length < 8,'7-7.9','> 8'))))
temp3=df1['sepal_length_cat'].value_counts().reset_index()
temp3['index']=pd.Categorical(temp3['index'],ordered=True,
                              categories=['<5','5-5.9','6-6.9','7-7.9','>8'])
```

```python
plt.bar(temp3.index,temp3.sepal_length_cat,color='green');
```

# Area Plot

```python
df4.head()
```

```python
df4['Country'].unique() #MENCARI VALUE YANG UNIK DARI SUATU KOLOM
```

```python
germany = df4[df4['Country'] == 'Germany'][['Year', 'Spending_USD']] #Menjadikan Country dan Tahun serta spending_usd sebagai variabel tersendiri
germany = germany.set_index('Year', drop=True) #Menjadikan Year Sebagai Index

france = df4[df4['Country'] == 'France'][['Year', 'Spending_USD']]
france = france.set_index('Year', drop=True)

great_britain = df4[df4['Country'] == 'Great Britain'][['Year', 'Spending_USD']]
great_britain = great_britain.set_index('Year', drop=True)

japan = df4[df4['Country'] == 'Japan'][['Year', 'Spending_USD']]
japan = japan.set_index('Year', drop=True)

usa = df4[df4['Country'] == 'USA'][['Year', 'Spending_USD']]
usa = usa.set_index('Year', drop=True)

canada = df4[df4['Country'] == 'Canada'][['Year', 'Spending_USD']]
canada = canada.set_index('Year', drop=True)
```

```python
temp3 = pd.concat([germany, france, great_britain, japan, usa, canada], axis=1)
temp3.columns = ['Germany', 'France', 'Great Britain', 'Japan', 'USA', 'Canada']
```

```python
# plt.figure() 
temp3.plot.area(figsize=(12,5))

plt.show()
```
